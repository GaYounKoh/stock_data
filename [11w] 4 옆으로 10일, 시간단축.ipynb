# mk_cd 함수

```python
def mk_cd(x):
    if type(x) == str:
        if x.isalpha():
            return x
        else:
            if len(x) == 6:
                return x
            else:
                return "0"*(6-len(x))+x
    elif type(x) == int:
        x = str(x)
        return "0"*(6-len(x))+x
```

```python
조건df['code'] = list(map(mk_cd, 조건df['code']))
```

# 10개 날짜씩 옆으로 잇기 csv 파일 포기, txt파일로...
* ten_rows 만들 때 앞 날짜(d-9)의 code가 d-0의 code와 같을 때라는 조건
* 원래 조건만족cd_dt.txt에 있던 row들 NaN값 있는 row 제거하면서 많이 사라짐...

```python
nc_cols = nc_df.columns.values
cols = []
for i in range(10):
    tmp_cols = [f'D-{9-i}_{col}' for col in nc_cols]
    cols.extend(tmp_cols)
len(cols)
```

# 단축 방법... dic을 만들어서 조건 확인
조건cd_dt.txt로 dic = {cd:[dt...]}을 만들어서 조건에 맞는 cddt 확인. <br>
nc_df 자체를 readlines 해줘야 시간 단축됨.

```python
조건dic = {j:[] for j in 조건df['code'].unique()}
for iter, i in tqdm(enumerate(l)):
    i = i.strip()
    i = i.split('\t') # 리스트에 문자열로 들어감

    
    cd = i[0]
    dt = i[1]
    if iter == 0:
        pass
    else:
        dt = int(dt)
        조건dic[cd].append(dt)

len(조건dic)
```

## nc_df readlines
* df 포기했고 txt 생성하기 위함.

nc_df에는 cd:1, dt:0 <br>
조건cd_dt에는 cd:0, dt:1 <br>

```python
f_nc = open('nc추가.csv', 'r', encoding = 'cp949')
l_nc = f_nc.readlines()
```

```python
# l_nc[iter].strip().split(',')
```

```python
all_lst = [cols]

for iter, i in tqdm(enumerate(l_nc)):
    i = i.strip()
    i = i.split(',') # 리스트에 문자열로 들어감
#     i
#     break
    
    cd = i[1]
    dt = i[0]
    if iter == 0:
        pass
    else:
        dt = int(dt)

        # code일 때 date인게 존재하면 조건idx 뽑는 코드 실행
        if dt in 조건dic[cd]:
            # 각 조건 괄호로 묶어줘야 결과 출력됨.
            
            # d-9 code == d-0 code일 때
            if l_nc[iter-9].strip().split(',')[1] == cd:
                ten_lst = l_nc[iter-9:iter+1]

            tmp = []
            for j in ten_lst:
                t = j.strip().split(',')

                # cols 바꾸기 단계 필요 없어짐.
                # index reset 단계도 필요 없어짐.
                
                tmp.append(t)

#             len(tmp)
            all_lst.append(tmp)
    #         all_lst
            

f.close()
```

# txt 저장

```python
with open(f'd9d0.txt', 'w', encoding = 'UTF-8') as f:
    for iter, lst in enumerate(all_lst):
        if iter == 0:
            tmp_l = ','.join(lst)
            tmp_l = tmp_l.replace(',', '\t')
            _ = f.write(f'{tmp_l}\n')
#         elif iter > 2:
#             break
        else:
            for i in lst:
                for j in i:
                    _ = f.write(f'{j}\t')
            _ = f.write('\n')
```

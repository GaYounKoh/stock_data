# data to csv

## new_cols

```python
nc_df = pd.read_feather('nc추가.ftr', columns = None, use_threads = True)
nc_df['code'].nunique()
# nc_df

nc_cols = nc_df.columns.values
cols = []
for i in range(10):
    tmp_cols = [f'D-{9-i}_{col}' for col in nc_cols]
    cols.extend(tmp_cols)
len(cols)
```

```python
new_cols = []
for i in cols:
    i = i.replace(', ','_') # 문제를 일으킬 수 있는 comma를 언더바로 대체
    print(f'{i},')
    new_cols.append(i)
# print(f'\n')
len(new_cols)
```

## d9d0.txt 로드

```python
f = open('d9d0.txt', 'r', encoding = 'cp949')
lines = f.readlines()
```

```python
for iter, i in tqdm(enumerate(lines)):
    i = i.strip()
#     i = i.split('\t')
    i = i.replace('\t', ',')
    len(i.split(',')) # 이미 colname에 ','가 자체적으로 붙어있는게 있어서... 그것이 길이에 문제가 된 것...
                        # tab으로 저장한 이유가 그거였음...
    i.split(',')[9:19]
#     print(f'{i}\n')
    if iter > 1:
        break
```

## data to csv (저장)

```python
with open(f'd9d0.csv', 'w', encoding = 'UTF-8') as f:
    for iter, i in tqdm(enumerate(lines)):
        
        # colnames 따로 설정
        if iter == 0:
            for j in new_cols:
                if j == new_cols[-1]:
                    _ = f.write(f'{j}')
                else:
                    _ = f.write(f'{j},')
            _ = f.write(f'\n')
        
        # 나머지는 그대로
        else:
            i = i.strip()
            i = i.replace('\t', ',')
            _ = f.write(f'{i}\n')
```

## df_cols 구성 변경
* 'D-0_code', 'D-0_date', ..., 'D-0_next_change'
** 나머지 날에 대해 code, date, next_change 제거.
* remove는 요소 자체로 array 내의 값 삭제 (해당 요소가 여러개면 제일 앞에 것 하나만 삭제됨.
* del은 요소의 idx로 array 내의 값 삭제.

```python
df_cols = list(df.columns.values)
len(df_cols)
```

```python
for i in range(9,-1,-1):
    DD = f'D-{i}_'
    df_cols.remove(DD+'date')
    df_cols.remove(DD+'code')
    df_cols.remove(DD+'next_change')
len(df_cols)
```

```python
new_df_cols = ['D-0_code', 'D-0_date'] + df_cols + ['D-0_next_change']
```

```python
new_df = df[new_df_cols]
```

## col 구성, 순서, col name 변경해서 csv 저장

```python
new_df.columns = ['code', 'date'] + df_cols + ['next_change']
```

```python
# csv 파일 새로 저장
new_df.to_csv('fin_df.csv', index = False)
```

# library

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

import pandas as pd
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

# data

```python
scaled = pd.read_feather('scaled.ftr', columns = None, use_threads = True)
```

```python
보조지표df = pd.read_feather('보조지표추가_cd_nuniq=2348.ftr', columns = None, use_threads = True)
```

# 사용할 cols
* (ma_60, ma_120지움, NaN값있는 부분 다 지우려고 하는데 ma_60이나 ma_120 있으면 데이터 너무 많이 지우게 돼서)

```python
cols = list(보조지표df.columns.values)
del cols[cols.index('ma_60')]
del cols[cols.index('ma_120')]
len(cols)
```

# ma_20의 NaN값이 있는 구간 지우기
* 해당 구간은 코드별로 앞의 19개구간

```python
df_all = pd.DataFrame()
for i in tqdm(보조지표df['code'].unique()):
    ## 종목코드 별 df
    tmp_df = scaled.iloc[np.where(scaled['code'] == i)]
    tmp19 = tmp_df[19:][cols]
    df_all = pd.concat([df_all, tmp19], ignore_index = True)
df_all
```

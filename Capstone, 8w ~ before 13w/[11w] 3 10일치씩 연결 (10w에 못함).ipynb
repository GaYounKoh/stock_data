# mk_cd 함수

```python
def mk_cd(x):
    if type(x) == str:
        if x.isalpha():
            return x
        else:
            if len(x) == 6:
                return x
            else:
                return "0"*(6-len(x))+x
    elif type(x) == int:
        x = str(x)
        return "0"*(6-len(x))+x
```

```python
조건df['code'] = list(map(mk_cd, 조건df['code']))
```

# 조건만족 cddt에 해당하는 cd dt와 일치하는 data에 대해 그거 포함 앞 10개 날짜의 data 뽑기

# 10개 날짜씩 옆으로 잇기
* ten_rows 만들 때 앞 날짜(d-9)의 code가 d-0의 code와 같을 때라는 조건
* 원래 조건만족cd_dt.txt에 있던 row들 NaN값 있는 row 제거하면서 많이 사라짐...

```python
all_df = pd.DataFrame()

for iter, i in tqdm(enumerate(l)):
    i = i.strip()
    i = i.split('\t') # 리스트에 문자열로 들어감

    
    cd = i[0]
    dt = i[1]
    if iter == 0:
        pass
    else:
        dt = int(dt)
#         cd, dt
#         (nc_df['code'] == cd).sum()
#         (nc_df['date'] == dt).sum()

        # code일 때 date인게 존재하면 조건idx 뽑는 코드 실행
        if (nc_df[nc_df['code'] == cd]['date']==dt).sum() == 1: # np.where를 안써도 충분히 조건에 맞는 행 찾는 거 가능했음...
            # 각 조건 괄호로 묶어줘야 결과 출력됨.
            조건idx = nc_df.iloc[np.where((nc_df['code'] == cd) & (nc_df['date'] == dt))].index.values[0]
        
        # d-9 code == d-0 code일 때
        if nc_df.iloc[조건idx-9]['code'] == nc_df.iloc[조건idx]['code']:
            ten_rows = nc_df.iloc[조건idx-9:조건idx+1]
            
        tmp = pd.DataFrame()
        for j in range(len(ten_rows)):
            t = ten_rows.iloc[[j]]
            
            # cols 바꾸기
            cols = [f'D-{9-j}_{col}' for col in t.columns]
#             cols
            t.columns = cols
#             t

            # index reset
            t = t.reset_index(drop = True)
            tmp = pd.concat([tmp, t], axis = 1)
#         tmp
        all_df = pd.concat([all_df, tmp], ignore_index = True)
#         all_df
            

f.close()
```

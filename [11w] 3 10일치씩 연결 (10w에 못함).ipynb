# mk_cd 함수

```python
def mk_cd(x):
    if type(x) == str:
        if x.isalpha():
            return x
        else:
            if len(x) == 6:
                return x
            else:
                return "0"*(6-len(x))+x
    elif type(x) == int:
        x = str(x)
        return "0"*(6-len(x))+x
```

```python
조건df['code'] = list(map(mk_cd, 조건df['code']))
```

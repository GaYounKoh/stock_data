# scaling by 전 날 종가

```python
df_all = pd.DataFrame()
for i in tqdm(보조지표df['code'].unique()):
    ## 종목코드 별 df
    tmp_df = 보조지표df.iloc[np.where(보조지표df['code'] == i)]
    
    # 각 종목코드의 맨 첫 날 tmp_df[0]은 기준일이므로 제외하고 생각
    # 각 종목코드의 맨 첫 날
    tmp0_가격관련지표 = tmp_df.iloc[[0]][가격관련지표]
    # tmp0_가격관련지표
    
    
    # tmp_df[1:][가격관련지표].reset_index().drop('index', axis = 1)/tmp_df['close'][:-1].reset_index().drop('index', axis = 1).values
    
    ## original df 바꿔보려고 했는데 안바뀜
    # tmp_df[1:][가격관련지표] = tmp_df[1:][가격관련지표].reset_index().drop('index', axis = 1)/tmp_df['close'][:-1].reset_index().drop('index', axis = 1).values # 24개 열
    # tmp_df[1:]
    ## 변수에 저장 (덮어쓰기)이 안되는 문제.... reset_index해서 그런 것 같음.
    # pd.concat([tmp0,tmp_df[1:]], ignore_index = True)
    
    tmp1_가격관련지표 = tmp_df[1:][가격관련지표].reset_index().drop('index', axis = 1)/tmp_df['close'][:-1].reset_index().drop('index', axis = 1).values
    tmp_가격only = pd.concat([tmp0_가격관련지표,tmp1_가격관련지표], ignore_index = True)
    
    ## 나머지 cols df, reset_index해서 쓰기
    tmp_else = tmp_df[cols].reset_index().drop('index', axis = 1)
    
    tmp_all = pd.concat([tmp_가격only, tmp_else], axis = 1)
    
    ## 아래로 이어붙이기, ignore_index하기
    df_all = pd.concat([df_all, tmp_all], ignore_index = True)

    ## 모든 열에 대해 나눠지는지 test
    # tmp_df[1:][['open']].reset_index().drop('index', axis = 1)/tmp_df['close'][:-1].reset_index().drop('index', axis = 1).values
    
    # reset_index하고 ignore_index옵션 켜주기
    # values을 붙여서 열 이름에 상관없게 해야 모든 열에 대해 나눠짐.
    # tmp_df[1:]['close'].reset_index().drop('index', axis = 1) / tmp_df['close'][:-1].reset_index().drop('index', axis = 1)
    
    
    ## 앞의 값과 NaN값 없이 1개 열로 잘 합쳐지는지 test
    # tmp = pd.DataFrame([1], columns = ['close']) # 'close'열의 맨 첫 줄은 기준이므로 1 혹은 그냥 값으로...
    # pd.concat([tmp, tmp_df[1:]['close'].reset_index().drop('index', axis = 1) / tmp_df['close'][:-1].reset_index().drop('index', axis = 1)], ignore_index = True)
    
    ## 값 잘나오는지 test
    # tmp_df[0:2]['close']
    # tmp_df[-2:]['close']
    # 37938 / 37588
#     tmp_df['close'][:-1]
#     break
df_all
```

```python
# 보조지표df['code'].unique()[-1]
tt = 보조지표df.iloc[np.where(보조지표df['code'] == 보조지표df['code'].unique()[-1])]
t0 = tt.iloc[[0]][가격관련지표]
t1 = tt[1:][가격관련지표].reset_index().drop('index', axis = 1)/tt['close'][:-1].reset_index().drop('index', axis = 1).values
t_가격only = pd.concat([t0,t1], ignore_index = True)
# t_가격only
t_else = tmp_df[cols].reset_index().drop('index', axis = 1) # 위에서 reset_index()도 하고 ignore_index도 켰기 때문에
                                                            # reset_index()해야함.
# t_else
t_all = pd.concat([t_가격only, t_else], axis = 1)
t_all
# d_all = pd.concat([df_all, tmp_all])
```

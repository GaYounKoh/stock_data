## df의 'code'열에 mk_cd 함수 적용.
* 모두 문자로 만들어 줄 것. (형태 획일화)

```python
def mk_cd(x):
    if type(x) == str:
        if x.isalpha():
            return x
        else:
            if len(x) == 6:
                return x
            else:
                return "0"*(6-len(x))+x
    elif type(x) == int:
        x = str(x)
        return "0"*(6-len(x))+x
```

```python
df['code'] = list(map(mk_cd, df['code'])) # map(함수, 적용할 데이터) 순으로 작성.

# 참고 https://blockdmask.tistory.com/531
```

## 본격 생성 cd

```python
ma = [5,20,60,120]
ma_df = pd.DataFrame()
for i in tqdm(조건df['code'].unique()):
    if i in ['002311', '006281', '007391', '010581', '015054', '015361',
             '016802', '017551', '019301', '025451', '028052', '146320',
             '251120', '288980', '330730', '365900', '366030', '370090',
             '371950', '376930', '388720', '396300', '406760']: # 'ADX'가 뭐길래 실행이 자꾸 안되는 것인지... 간혹 'ATR'도 있음.
                        # 존속일수가 너무 짧아서 그런듯. 짧다의 기준: 20일전후
                        # 오류 뜨는 종목코드들은 일일이 찾음.
        continue
    else:
        tmp_df = df.iloc[np.where(df['code'] == i)]
#     tmp_df # 확인

    for days in ma: # day일 동안
        tmp_df['ma_'+str(days)] = tmp_df['close'].rolling(window = days).mean()
#         if len(tmp_df) > days:
#             tmp_df['ma_'+str(days)] = tmp_df['close'].rolling(window = days).mean()
#         else:
#             tmp_df['ma_'+str(days)] = 0
    
    ###### 병합 전에 보조지표 생성
    H, L, C, V = tmp_df['high'], tmp_df['low'], tmp_df['close'], tmp_df['volume']
    
    tmp_df['MFI'] = ta.volume.money_flow_index(high=H, low=L, close=C, volume=V, fillna=True)

    tmp_df['ADI'] = ta.volume.acc_dist_index(high=H, low=L, close=C, volume=V, fillna=True)

    tmp_df['OBV'] = ta.volume.on_balance_volume(close=C, volume=V, fillna=True)
    tmp_df['CMF'] = ta.volume.chaikin_money_flow(high=H, low=L, close=C, volume=V, fillna=True)

    tmp_df['FI'] = ta.volume.force_index(close=C, volume=V, fillna=True)
    tmp_df['EOM, EMV'] = ta.volume.ease_of_movement(high=H, low=L, volume=V, fillna=True)

    tmp_df['VPT'] = ta.volume.volume_price_trend(close=C, volume=V, fillna=True)
    tmp_df['NVI'] = ta.volume.negative_volume_index(close=C, volume=V, fillna=True)
    tmp_df['VMAP'] = ta.volume.volume_weighted_average_price(high=H, low=L, close=C, volume=V, fillna=True)

    # Volatility
    tmp_df['ATR'] = ta.volatility.average_true_range(high=H, low=L, close=C, fillna=True)
    tmp_df['BHB'] = ta.volatility.bollinger_hband(close=C, fillna=True)
    tmp_df['BLB'] = ta.volatility.bollinger_lband(close=C, fillna=True)
    tmp_df['KCH'] = ta.volatility.keltner_channel_hband(high=H, low=L, close=C, fillna=True)
    tmp_df['KCL'] = ta.volatility.keltner_channel_lband(high=H, low=L, close=C, fillna=True)
    tmp_df['KCM'] = ta.volatility.keltner_channel_mband(high=H, low=L, close=C, fillna=True)
    tmp_df['DCH'] = ta.volatility.donchian_channel_hband(high=H, low=L, close=C, fillna=True)
    tmp_df['DCL'] = ta.volatility.donchian_channel_lband(high=H, low=L, close=C, fillna=True)
    tmp_df['DCM'] = ta.volatility.donchian_channel_mband(high=H, low=L, close=C, fillna=True)
    tmp_df['UI'] = ta.volatility.ulcer_index(close=C, fillna=True)
    # Trend
    tmp_df['SMA'] = ta.trend.sma_indicator(close=C, fillna=True)
    tmp_df['EMA'] = ta.trend.ema_indicator(close=C, fillna=True)
    tmp_df['WMA'] = ta.trend.wma_indicator(close=C, fillna=True)
    tmp_df['MACD'] = ta.trend.macd(close=C, fillna=True)
    tmp_df['ADX'] = ta.trend.adx(high=H, low=L, close=C, fillna=True)
    tmp_df['-VI'] = ta.trend.vortex_indicator_neg(high=H, low=L, close=C, fillna=True)
    tmp_df['+VI'] = ta.trend.vortex_indicator_pos(high=H, low=L, close=C, fillna=True)
    tmp_df['TRIX'] = ta.trend.trix(close=C, fillna=True)
    tmp_df['MI'] = ta.trend.mass_index(high=H, low=L, fillna=True)
    tmp_df['CCI'] = ta.trend.cci(high=H, low=L, close=C, fillna=True)
    tmp_df['DPO'] = ta.trend.dpo(close=C, fillna=True)
    tmp_df['KST'] = ta.trend.kst(close=C, fillna=True)
    tmp_df['Ichimoku'] = ta.trend.ichimoku_a(high=H, low=L, fillna=True)
    tmp_df['Parabolic SAR'] = ta.trend.psar_down(high=H, low=L, close=C, fillna=True)
    tmp_df['STC'] = ta.trend.stc(close=C, fillna=True)
    # Momentum
    tmp_df['RSI'] = ta.momentum.rsi(close=C, fillna=True)
    tmp_df['SRSI'] = ta.momentum.stochrsi(close=C, fillna=True)
    tmp_df['TSI'] = ta.momentum.tsi(close=C, fillna=True)
    tmp_df['UO'] = ta.momentum.ultimate_oscillator(high=H, low=L, close=C, fillna=True)
    tmp_df['SR'] = ta.momentum.stoch(close=C, high=H, low=L, fillna=True)
    tmp_df['WR'] = ta.momentum.williams_r(high=H, low=L, close=C, fillna=True)
    tmp_df['AO'] = ta.momentum.awesome_oscillator(high=H, low=L, fillna=True)
    tmp_df['KAMA'] = ta.momentum.kama(close=C, fillna=True)
    tmp_df['ROC'] = ta.momentum.roc(close=C, fillna=True)
    tmp_df['PPO'] = ta.momentum.ppo(close=C, fillna=True)
    tmp_df['PVO'] = ta.momentum.pvo(volume=V, fillna=True)
    ######
    
    
    ma_df = pd.concat([ma_df,tmp_df])
    
#     break
# tmp_df
ma_df
```

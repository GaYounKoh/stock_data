```python
df = pd.read_csv('stock_2018.csv')
df = df.iloc[:,1:] # index열 제거
df.sort_values(['Code', 'Date'], inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df_all = pd.concat([df, cci['cci']], axis = 1)
```

```python
## 일단 아이디어는 이럼. 이걸 open으로 해결하기.
# for i in range(len(df_all)):
#     if df_all['BLB'][i] <= df_all['Close'] <= df_all['BHB'][i]:

touch = []

f = open('df_all.csv')
lines = f.readlines()
for i,line in tqdm(enumerate(lines)):
    line = line.strip()
    lst = line.split(',')
    if i == 0:
        close = lst.index('Close')
        blb = lst.index('BLB')
        bhb = lst.index('BHB')
        cci = lst.index('cci')
    
    else:
        if (lst[blb] <= lst[close] <= lst[bhb]) and (-20 <= float(lst[cci]) <= 20):
            touch.append(1)
        else:
            touch.append(0)
f.close()
```

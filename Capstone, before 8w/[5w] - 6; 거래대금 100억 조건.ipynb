```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv('add_BB_touch.csv')
df
```

```python
idx_lst = [] # tv >=100억인 인덱스를 뽑아놓고 그보다 앞 10개는 나중에 이 list를 for문 돌려서 앞 10개 뽑는 식으로 진행.

f = open('add_BB_touch.csv')
lines = f.readlines()
for i,line in tqdm(enumerate(lines)):
    line = line.strip()
    lst = line.split(',')
    if i == 0:
        tv = lst.index('Trading_Value')
#         hi = lst.index('High')
#         lo = lst.index('Low')
#         close = lst.index('Close')
#         blb = lst.index('BLB')
#         bhb = lst.index('BHB')
#         cci = lst.index('cci')
    else:
        if int(lst[tv]) >= 100_0000_0000:
            idx_lst.append(i-1)
#         break
```

```python
len(idx_lst)
```

```python
## 해당일자를 포함한 d-9 ~ d-0 : x
## 그리고 d-0의 next change : y

dic = {} # 일단 계속 idx를 뽑아두는 중... 형태는 d-0의 idx : d-9 ~ d-0의 idx list

for i in tqdm(df['Code'].unique()):
    a = set(df[df['Code'] == i].index) & set(idx_lst) # 둘이 겹치는 부분이 있다 == 'Code'==i 일때의 idx 중에 tv>=100억인 idx가 있다.
#     list(df[df['Code'] == i].index).index(33)
    for j in list(a):
        if list(df[df['Code'] == i].index).index(j) > 9:
            dic[j] = list(range(j-9,j+1))
#     break
# dic
```

# 100억 row들 뽑아서 옆으로, 아래로 잇기

```python
# dic 개수 38054밖에 안되는데 왜이렇게 오래걸림...
dday_alldf = pd.DataFrame()
for i in tqdm(dic.keys()): # 한 key안에 있는 애들은 전부 옆으로 붙이기
    for k,j in enumerate(dic[i]):
        if k == 0:
            tmp0 = df.iloc[j,2:]
            tmp0.index = [f'D-{9-k}_{col}' for col in tmp0.index]
            tmp0 = pd.DataFrame(tmp0).T
            tmp0 = tmp0.rename(index = {j : i})
        elif k == 9:
            tmp_idx = df.loc[[j]][['Code', 'Date']]
            tmp = df.iloc[j,2:]
            tmp.index = [f'D-{9-k}_{col}' for col in tmp.index]
            tmp = pd.DataFrame(tmp).T
            tmp = tmp.rename(index = {j : i})
            tmp0 = pd.concat([tmp_idx, tmp0, tmp], axis = 1)
            
        else:
#             df.loc[[j]]
            tmp = df.iloc[j,2:]
            tmp.index = [f'D-{9-k}_{col}' for col in tmp.index]
            tmp = pd.DataFrame(tmp).T
            tmp = tmp.rename(index = {j : i})
            tmp0 = pd.concat([tmp0, tmp], axis = 1)
        
    dday_alldf = pd.concat([dday_alldf, tmp0])

dday_alldf
```

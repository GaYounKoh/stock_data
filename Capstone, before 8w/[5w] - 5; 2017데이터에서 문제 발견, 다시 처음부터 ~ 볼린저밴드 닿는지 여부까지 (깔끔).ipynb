```python
f = open(f'data2017with138250.txt', 'r', encoding = 'utf-8')
lines = f.readlines()
df2017 = pd.DataFrame()
for line in lines:
    cd = line.strip()
    tmp = fdr.DataReader(cd, start = '2017-01-01', end = '2017-12-31')

    tmp_cd = pd.DataFrame([cd] * len(tmp))# pd.Series([cd] * len(tmp), name = 'Code')
    tmp_cd.index = tmp.index # 원랜 안해줘도 됐는데 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~.
    tmp_cd.columns = ['Code'] # 원랜 안해줘도 됐는데 이번엔 안해줬더니 이상하게 받길래... (axis 문제인지 확인해봤지만 아닌듯 보였음.)
    
    # 옆으로 이어붙이기
    fin_tmp = pd.concat([tmp_cd, tmp], axis = 1) # 아니 axis = 1 안했을 때 왜 옆으로 들어간 것처럼 보여선 헷갈리게...
    
    # 아래로 이어붙이기, 이름 덮어써야 계속 이어붙음.
    df2017 = pd.concat([df2017, fin_tmp])
f.close()
```

# 인터넷 코드로 적용해보기. (re_s_all)
* s_all 만들고 M 만들고 바로 넘어오기... <br>
* 같은 code일 때만 묶어서 구글링 코드 적용시킬것...

```python
re_s_all = s_all.copy()
```

```python
re_s_all['Code'].unique()
re_s_all['Code'].nunique()
```

```python
sma = [] # m
adv = [] # d
cci = [] # cci

for i in re_s_all['Code'].unique():
    tmp_sma = re_s_all[re_s_all['Code'] == i]['M'].rolling(20).mean()
    tmp_adv = re_s_all[re_s_all['Code'] == i]['M'].rolling(20).apply(lambda x:pd.Series(x).mad()) # MAD: mean absolute deviation는 평균 절대 편차
    tmp_cci = (re_s_all[re_s_all['Code'] == i]['M'] - tmp_sma) / (0.015 * tmp_adv)
    sma.extend(list(tmp_sma.values))
    adv.extend(list(tmp_adv.values))
    cci.extend(list(tmp_cci.values))

len(sma) # 셋 다 461795
len(adv)
len(cci)

# re_s_all
```

```python
re_s_all_again = pd.concat([re_s_all, pd.Series(sma, name = 'm')], axis = 1)
re_s_all_again = pd.concat([re_s_all_again, pd.Series(adv, name = 'd')], axis = 1)
re_s_all_again = pd.concat([re_s_all_again, pd.Series(cci, name = 'cci')], axis = 1)
```

# 각 hi, lo가 각 BB +- 20 이내에 있는지 여부 열 생성

```python
add_cci_range = pd.read_csv('add_cci_range.csv')
```

```python
## 아이디어: 과거 lo 혹은 hi 와 현재 lo 혹은 hi 사이에 현재 blb 혹은 bhb가 있으면 0.5 또는 1 추가

touch = []

f = open('df_all_fin.csv')
lines = f.readlines()
for i,line in tqdm(enumerate(lines)):
    line = line.strip()
    lst = line.split(',')
    if i == 0:
        hi = lst.index('High')
        lo = lst.index('Low')
        close = lst.index('Close')
        blb = lst.index('BLB')
        bhb = lst.index('BHB')
        cci = lst.index('cci')
    
    else:
        if i == 1:
            tmp_hi = lst[hi]
            tmp_lo = lst[lo]
            
            if float(lst[bhb])-20 <= float(lst[hi]) <= float(lst[bhb])+20:
                touch.append(1)

            elif float(lst[blb])-20 <= float(lst[lo]) <= float(lst[blb])+20:
                touch.append(0.5)

            else:
                touch.append(0)
                
        else:
            if tmp_hi <= lst[bhb] <= lst[hi]:
                touch.append(1)

            elif tmp_lo <= lst[blb] <= lst[lo]:
                touch.append(0.5)

            else:
                touch.append(0)
                
            tmp_hi = lst[hi] # 계속 업데이트
            tmp_lo = lst[lo]
f.close()
```

```python
(np.array(touch)==1).sum()
(np.array(touch)==0.5).sum()
(np.array(touch)==0).sum()
```

```python
add_BB_touch = pd.concat([add_cci_range, pd.Series(touch, name = 'BB_only')], axis = 1)
```

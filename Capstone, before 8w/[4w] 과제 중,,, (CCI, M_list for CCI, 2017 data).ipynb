## code 담긴 file load 해서 df 하나로 합치기 (df2017)

```python
f = open(f'data2017.txt', 'r', encoding = 'utf-8')
lines = f.readlines()
df2017 = pd.DataFrame()
for line in lines:
    cd = line.strip()
    tmp = fdr.DataReader(cd, start = '2017-01-01', end = '2017-12-31')
#     tmp
#     tmp.index
    tmp_cd = pd.DataFrame([cd] * len(tmp))# pd.Series([cd] * len(tmp), name = 'Code')
    tmp_cd.index = tmp.index # 원랜 안해줘도 됐는데 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~.
    tmp_cd.columns = ['Code'] # 원랜 안해줘도 됐는데 이번엔 안해줬더니 이상하게 받길래... (axis 문제인지 확인해봤지만 아닌듯 보였음.)
#     tmp_cd
    
    # 옆으로 이어붙이기
    fin_tmp = pd.concat([tmp_cd, tmp], axis = 1) # 아니 axis = 1 안했을 때 왜 옆으로 들어간 것처럼 보여선 헷갈리게...
    
    # 아래로 이어붙이기, 이름 덮어써야 계속 이어붙음.
    df2017 = pd.concat([df2017, fin_tmp])
```

## 2017 M_lst (M_lst는 cci지수 만들 때 필요한 것.)

```python
M_lst2017 = []

f = open('df2017.csv')
lines = f.readlines()
for i, line in tqdm(enumerate(lines)):
    line = line.strip()
    lst = line.split(',')
    
    if i == 0:
        hi = lst.index('High') # 3
        lo = lst.index('Low') # 4
        cl = lst.index('Close') # 5
    else:
        M = 0
        for i in (hi,lo,cl):
#             'i', i, 'lst[i]:', lst[i]
            M += int(lst[i])
        M_lst2017.append(M/3)
```

## 2017 CCI

```python
cci2017 = pd.DataFrame(M_lst2017, columns = ['AP']) #M
cci2017['SMA'] = cci2017['AP'].rolling(20).mean() #m
cci2017['ADV'] = cci2017['AP'].rolling(20).apply(lambda x:pd.Series(x).mad()) #d
cci2017['CCI'] = (cci2017['AP'] - cci2017['SMA']) / (0.015 * cci2017['ADV']) #cci
cci2017
```
